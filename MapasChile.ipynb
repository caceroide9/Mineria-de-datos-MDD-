```python
def rd2(x):
    return round(x, 2)


minimum, maximum = regiones["COD_REGI"].quantile([0.05, 0.95]).apply(rd2)

mean = round(regiones["COD_REGI"].mean(), 2)


print(f"minimum: {minimum}", f"maximum: {maximum}", f"Mean: {mean}", sep="\n\n")
```

```python
import branca


colormap = branca.colormap.LinearColormap(
    colors=["#f2f0f7", "#cbc9e2", "#9e9ac8", "#756bb1", "#54278f"],
    index=regiones["COD_REGI"].quantile([0.2, 0.4, 0.6, 0.8]),
    vmin=minimum,
    vmax=maximum,
)

colormap.caption = "Population Density in the United States"

colormap
```

```python
ch_comunas = geopandas.sjoin(comunas, regiones, how="inner", predicate="within")

pop_ranked_cities = ch_comunas.sort_values(by="CODIGO", ascending=False)
```

```python
import folium
from folium.plugins import Search


m = folium.Map(location=[-34.3262849,-71.2447849],zoom_start=5)


def style_function(x):
    return {
        "fillColor": colormap(x["properties"]["COD_REGI"]),
        "color": "black",
        "weight": 2,
        "fillOpacity": 0.5,
    }
```

```python
regiongeo = folium.GeoJson(
    regiones,
    name="US States",
    style_function=style_function,
    tooltip=folium.GeoJsonTooltip(
        fields=["NOM_REG", "COD_REGI"], aliases=["NOM_REG", "COD_REGI"], localize=True
    ),
).add_to(m)




comunageo = folium.GeoJson(
    pop_ranked_cities,
    name="US Cities",
    tooltip=folium.GeoJsonTooltip(
        fields=["NOM_COM", "CODIGO"], aliases=["NOM_COM", "CODIGO"], localize=True
    ),
).add_to(m)
```

```python
statesearch = Search(
    layer=regiongeo,
    geom_type="Polygon",
    placeholder="Buscar Regi贸n",
    collapsed=False,
    search_label="NOM_REG",
    weight=3,
).add_to(m)

citysearch = Search(
    layer=comunageo,
    geom_type="Point",
    placeholder="Buscar Comuna",
    collapsed=True,
    search_label="NOM_COM",
).add_to(m)

```
